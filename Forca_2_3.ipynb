```python
###Versao 2.3
###Criada por José Ivan G Silva
##Temos agora um arquivo externo que pode-se colocar varias palavras
###no codigo de exemplo foi colocado arquivo do google drive via colabs
###acrescentada e melhorada a rotinaque exibe as letras digitadas e o painel de acertos
'''
Jogo da forca :Proposto pelo curso da DataScienceAcademy
'''
import random

###Imprime a figura correspondente da forca
#3Vai aumentando até dar Game Over.
##GAME OVER == range(len(figura)) -- comprimento é o tamanho da palavra sorteada
def desenha(index):
  figura = ['''

>>>>>>>>>>FORCA 2.2 <<<<<<<<<<

+---+
|   |
    |
    |
    |
    |
=========''', '''

+---+
|   |
O   |
    |
    |
    |
=========''', '''

+---+
|   |
O   |
|   |
    |
    |
=========''', '''

 +---+
 |   |
 O   |
/|   |
     |
     |
=========''', '''

 +---+
 |   |
 O   |
/|\  |
     |
     |
=========''', '''

 +---+
 |   |
 O   |
/|\  |
/    |
     |
=========''', '''

 +---+
 |   |
 O   |
/|\  |
/ \  |
     |
=========''']
  print(figura[index])
###Função para gerar um sorteio em uma lista de palavras.
###vai retornar uma lista de caracteres, que é o mesmo que uma string
###exemplo de utilização : print(palavraAleatoria([AQUI VAI UMA LISTA DE PALAVRAS]))
def palavraAleatoria(ListaPalavras):  
  a =random.randrange(0,len(ListaPalavras))
  sorteio = ListaPalavras[a]
  matriz= []
  letrasDigitadas = ['0']
  resultado = False

  for i in range(len(sorteio)):
    matriz.append(sorteio[i])
  return matriz

###A FUNÇÃO QUE TRANSFORMA  O ARQUIVO TXT EM UMA LISTA DE PALAVRAS
def transforma_lista(arquivo):
  with open(arquivo, "r") as arquivo:
    lines = arquivo.readlines()
    lines = list(map(lambda s: s.strip(), lines))  
    arquivo.close()  
    return lines

##CHAMA A FUNCAO QUE VAI TRANFORMAR EM LISTA 
arquivo_lista = transforma_lista("/content/drive/MyDrive/Arquivo_txt/palavras_forca.txt")

###CHAMA A FUNCAO QUE SORTEIA UMA PALAVRA DA LISTA
matriz = palavraAleatoria(arquivo_lista) 
esc = ' '
pontos = 0
erro = 0
placa = []
digitados = []
vitoria = False

for i in range(len(matriz)):
  placa.append('_ ')

print('DICA IMPORTANTE : Tem ',len(matriz,),' letras.')

while esc != '0':
###DESCOMENTE A LINHA ABAIXO SE QUISER VER A PALAVRA SORTEADA E EXECUTAR TESTES NO CÓDIGO
  ##print(matriz)
  esc =input('Digite : ')  
  
  if matriz.count(esc)==0:
    if digitados.count(esc) == 0:
        digitados.append(esc)  
        desenha(erro)
        if erro == 6:
          esc ='0'
        else:
          print(digitados)
          erro+=1
    else:
        print('Voce ja digitou esta letra.')

        
  elif matriz.count(esc) > 0:
    if digitados.count(esc) == 0:
      pontos = pontos +matriz.count(esc)
      digitados.append(esc)      
      print('Acertos : ' ,pontos)

      for j in range(len(matriz)):
        if matriz[j] == esc:
          placa[j] = esc

      ###EXIBINDO LETRAS ESCONDIDAS ,SE ACERTAR VAI APARECENDO
      for k in placa:
        print(k , end = ' ')

      if pontos == len(matriz):      
        vitoria = True
        esc = '0'
    else:
      print(digitados)
      print('Você já digitou este numero')
      
if vitoria:
  print('Voce Ganhou!!', end = '  ') 
  print('A palavra é ', end=' ')
  for j in matriz:
    print(j, end=' ')
else:
  print('Voce Perdeu!!') 
  print('Pontos : ', pontos)
  print('Erros' ,erro + 1)
```

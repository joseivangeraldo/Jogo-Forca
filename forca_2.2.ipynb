```python

####para exibir letras na palavra secreta
###ex _ _ _ _ palavra bala.Conforme
#vai acertando preenche a secreta
'''
Jogo da forca :Proposto pelo curso da DataScienceAcademy
'''
import random

###Imprime a figura correspondente da forca
#3Vai aumentando até dar Game Over.
##GAME OVER == range(len(figura)) -- comprimento é o tamanho da palavra sorteada
def desenha(index):
  figura = ['''

>>>>>>>>>>FORCA 2.2 <<<<<<<<<<

+---+
|   |
    |
    |
    |
    |
=========''', '''

+---+
|   |
O   |
    |
    |
    |
=========''', '''

+---+
|   |
O   |
|   |
    |
    |
=========''', '''

 +---+
 |   |
 O   |
/|   |
     |
     |
=========''', '''

 +---+
 |   |
 O   |
/|\  |
     |
     |
=========''', '''

 +---+
 |   |
 O   |
/|\  |
/    |
     |
=========''', '''

 +---+
 |   |
 O   |
/|\  |
/ \  |
     |
=========''']
  print(figura[index])
###Função para gerar um sorteio em uma lista de palavras.
###vai retornar uma lista de caracteres, que é o mesmo que uma string
###exemplo de utilização : print(palavraAleatoria([AQUI VAI UMA LISTA DE PALAVRAS]))
def palavraAleatoria(ListaPalavras):  
  a =random.randrange(0,len(ListaPalavras))
  sorteio = ListaPalavras[a]
  matriz= []
  letrasDigitadas = ['0']
  resultado = False

  for i in range(len(sorteio)):
    matriz.append(sorteio[i])
  return matriz

###A MATRIZ QUE VAI PEGAR A FUNÇÃO palavraAleatoria E SPRTEAR UMA DAS PALAVRAS
matriz = palavraAleatoria(['bala' , 'casa' , 'doce','rato']) 
esc = ' '
pontos = 0
erro = 0
placa = []
digitados = []
vitoria = False

for i in range(len(matriz)):
  placa.append('_ ')

print('DICA IMPORTANTE : Tem ',len(matriz,),' letras.')


while esc != '0':
  esc =input('Digite : ')
  for i in range(len(matriz)):
    if matriz[i] == esc:
      placa[i] = esc
  print(placa)
  
  if matriz.count(esc)==0:
    digitados.append(esc)    
    desenha(erro)
    if erro == 6:
      esc ='0'
    else:
      print(digitados)
      erro+=1
        
  elif matriz.count(esc) > 0:
    if digitados.count(esc) == 0:
      pontos = pontos +matriz.count(esc)
      digitados.append(esc)
      print('Acertos : ' ,pontos)
      if pontos == len(matriz):      
        vitoria = True
        esc = '0'
    else:
      print(digitados)
      print('Você já digitou este numero')
      
if vitoria:
  print('Voce Ganhou!!') 
else:
  print('Voce Perdeu!!') 
  print('Pontos : ', pontos)
  print('Erros' ,erro + 1)
```
